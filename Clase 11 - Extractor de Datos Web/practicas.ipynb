Extraemos imagenes

```python
# Extraemos imagenes
# Obtenemos la pagina con requests.get
resultado_img = requests.get('https://escueladirecta.com/')
```

```python
sopa_img = bs4.BeautifulSoup(resultado_img.text, 'lxml')
```

```python
imagenes = sopa_img.select('.course-box-image')
#imagenes
```

```python
# Obtenemos la url de cada imagen
lista_url = []
for i in imagenes:
    lista_url.append(i['src'])
```

```python
j= 1
for i in lista_url:
    res_imagen = requests.get(i) # traemos informacion de la url de la imagen
    
    # Creamos y guardamos la imagen en el ordenador
    f = open(f'./imagenes/imagen_{j}.jpg', 'wb')
    f.write(res_imagen.content)
    f.close()
    j += 1
```
